```python
# Función para verificar y reemplazar el precio por el valor que corresponda.
def verifica_y_reemplaza(valor):
    if isinstance(valor, str):
        if 'free' in valor.lower():
            return 0.0
        elif 'Starting at $499.00' in valor:
            return 499.0
        elif 'Starting at $449.00' in valor:
            return 449.0
        else:         
            return None
    else:
        return valor

# Corregimos los valores aplicando nuestra funcion
games["price"] = games["price"].apply(verifica_y_reemplaza)
```

```python
# Funcion para leer un archivo JSON corrupto y transformarlo a uno valido (Lo usaremos con "user reviews" y con "user items")

def convertir_json(ruta: str, atributo_to_fix: str, detector_fin: str):
    lista_auxiliar = []
    with open(ruta, encoding = 'utf-8') as f:
        for line in f:
            while f"'{atributo_to_fix}': '" in line or f"'{atributo_to_fix}': \"" in line:
                if f"'{atributo_to_fix}': '" in line:
                    inicio = f"'{atributo_to_fix}': '"
                    fin = f"'{detector_fin}"
                elif f"'{atributo_to_fix}': \"" in line:
                    inicio = f"'{atributo_to_fix}': \""
                    fin = f"\"{detector_fin}"
                #print(inicio)
                #print(fin)

                inicio_index = line.find(inicio)
                fin_index = line.find(fin, inicio_index + len(inicio))
                inicio_modificado = inicio.replace(" ", "*borrar*")
                #print(inicio_index)
                #print(fin_index)

                if inicio_index != -1 and fin_index != -1:
                    porcion_a_modificar = line[inicio_index + len(inicio):fin_index]
                    #print(porcion_a_modificar)
                    porcion_modificada = porcion_a_modificar.replace("\"", "**")
                    porcion_modificada = porcion_modificada.replace("'", "**")
                    porcion_modificada = porcion_modificada.replace("\\", "")
                    line = line[:inicio_index] + inicio_modificado + porcion_modificada + line[fin_index:]
                else:
                    break
            
            line = line.replace("\"", "'")
            line = line.replace("'", "\"")
            line = line.replace("**" , "'")
            line = line.replace("*borrar*", " ")
            line = line.replace("True", "true")
            line = line.replace("False", "false")
            #print(line[max(0, 1394 - 20):min(len(line), 1394 + 20)])
            fila = json.loads(line)
            lista_auxiliar.append(fila)

        return lista_auxiliar
```

```python
# Creamos una funcion para transformar una lista de diccionarios a varias columnas de un df

def desanidar_lista_dic(df: pd.DataFrame, columna: str, claves: list):
    serie = pd.Series(data = [])
    for clave in claves:
        for index, row in df.iterrows():
            lista_diccionarios = row[columna]
            lista = list()
            for diccionario in lista_diccionarios:
                if clave in diccionario:
                    lista.append(diccionario[clave])
            serie.loc[index] = lista
        df[clave] = serie
    return df
```

Para dos funciones que necesitamos para los endpoints de la API, vamos a generar un dataset auxiliar de "items" para luego servir los datos limpios y evitar consumo de recursos

```python
def desanidar_lista_dic_aux(df: pd.DataFrame, columna: str, claves: list):
    df_exploded = df.explode(columna)

    # Luego, crea columnas para cada clave
    for clave in claves:
        df_exploded[clave] = df_exploded[columna].apply(lambda x: x.get(clave) if isinstance(x, dict) else None)

    # Elimina la columna original de diccionarios
    df_exploded.drop(columna, axis=1, inplace=True)

    return df_exploded
```

```python
# Debemos transformar a una fecha valida el siguiente formato.
user_reviews["posted"][0]
```

```python
# Creamos una funcion para convertir al formato YYYY-MM-DD
def convertir_a_fecha(fecha):
    try:
        fecha_limpia = fecha.replace('Posted ', '').replace('.', '')
        fecha_obj = pd.to_datetime(fecha_limpia, format='%B %d, %Y', errors='coerce')
        if not pd.isna(fecha_obj):
            return fecha_obj.strftime('%Y-%m-%d')
    except:
        pass
    return None
```

```python
# Aplicamos la función a cada elemento de cada fila de "posted"
user_reviews["posted"] = user_reviews["posted"].apply(lambda lista: [convertir_a_fecha(fecha) for fecha in lista])
```

```python
# Estos columnas, que incluyen listas en sus elementos, las debe tomar literal como listas (no string)
user_reviews['posted'] = user_reviews['posted'].apply(ast.literal_eval)
user_reviews['item_id'] = user_reviews['item_id'].apply(ast.literal_eval)
user_reviews['recommend'] = user_reviews['recommend'].apply(ast.literal_eval)
user_reviews['review'] = user_reviews['review'].apply(ast.literal_eval)

# Hacemos un explode de los atributos deseados
user_reviews = user_reviews.explode(["posted", "item_id", "recommend", "review"])
```
